### Data Cleaning

Find all null values in the dataset. If there is any null values  in any column then fill it with mean of that column

```python
#check for null values
car.isnull().sum()
```

```python
#so now we have got the idea of null values
#we have relaced the null value by Omni 
car['Make'].fillna('Omni',inplace=True)
```

```python
car['Make'].isnull().sum()
```

```python
#similarly other columns are also be treated to remove null values
car['Model'].fillna('MDX',inplace=True)
```

```python
car['Type'].fillna('Sedan',inplace=True)
```

```python
car['Origin'].fillna('Asia',inplace=True)
```

```python
car['DriveTrain'].fillna('Front',inplace=True)
```

```python
car['MSRP'].fillna('$36,945',inplace=True)
```

```python
car['Invoice'].fillna('$33,337',inplace=True)
```

```python
car['EngineSize'].fillna(car['EngineSize'].mean(),inplace=True)
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(),inplace=True)
```

```python
car['Horsepower'].fillna(car['Horsepower'].mean(),inplace=True)
```

```python
car['MPG_City'].fillna(car['MPG_City'].mean(),inplace=True)
```

```python
car['MPG_Highway'].fillna(car['MPG_Highway'].mean(),inplace=True)
```

```python
car['Weight'].fillna(car['Weight'].mean(),inplace=True)
```

```python
car['Wheelbase'].fillna(car['Wheelbase'].mean(),inplace=True)
```

```python
car['Length'].fillna(car['Length'].mean(),inplace=True)
```

```python
#now all columns are well adjusted
car.isnull().sum()
```
